```python
import os
import cv2

import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split,GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report

from skimage.feature import hog, graycomatrix, graycoprops, local_binary_pattern
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.transform import resize
from imgaug import augmenters as iaa

from sklearn.ensemble import HistGradientBoostingClassifier as gb
from sklearn.naive_bayes import GaussianNB

# from imblearn.over_sampling import SMOTE
from concurrent.futures import ThreadPoolExecutor
```

```python
DATASET_CITRA = 'dataset'
IMG_SIZE = (64, 64)
AUGMENT_FACTOR = 5

# Parameter Ekstraksi Fitur
HOG_ORIENTATIONS = 9
LBP_POINTS = 24
GLCM_DISTANCES = [1, 2, 3]
GLCM_ANGLES = [0, 45, 90, 135]

# Parameter SVM
RANDOM_STATE = 42
TEST_SIZE = 0.2
```

### 1. Pre-processing Data

```python
augmentasi = iaa.Sequential([
    iaa.Fliplr(0.5),
    iaa.Affine(
        rotate=(-45, 45),
        shear=(-16, 16)
    ),
    iaa.GaussianBlur(sigma=(0, 0.5)),
    iaa.LinearContrast((0.75, 1.5)),
    iaa.AdditiveGaussianNoise(loc=0, scale=(0.0, 0.05*255), per_channel=0.5)
], random_order=True)

def augment_image(img):
    return augmentasi.augment_image(img)

# Visualisasi augmentasi
def visual_aug(img_path):
    img_lokasi = imread(img_path)
    plt.figure(figsize=(15, 5))

    plt.subplot(1,5,1)
    plt.imshow(img_lokasi)
    plt.title("Original")

    for i in range(4):
        plt.subplot(1,5,i+2)
        aug_img = augment_image(img_lokasi)
        plt.imshow(aug_img)
        plt.title(f"Augmented {i+1}")

    plt.tight_layout()
    plt.show()

lokasi = './dataset/Anthracnose/Anthracnose(1).jpg'
visual_aug(lokasi)
```

### Extraksi Fitur

```python
def proses_image(imag):
    if len(imag.shape) > 2:
        data_citra = rgb2gray(imag)
    else:
        data_citra = imag

    data_citra = resize(data_citra, (64, 64))
    gray = (data_citra * 255).astype(np.uint8)

    # HOG
    hog_features = hog(gray, orientations=HOG_ORIENTATIONS,
                      pixels_per_cell=(16, 16),
                      cells_per_block=(1, 1),
                      visualize=False)
    # lbp
    lbp = local_binary_pattern(gray, P=LBP_POINTS, R=3, method='uniform')
    hist_lbp, _ = np.histogram(lbp, bins=np.arange(0, LBP_POINTS+3), range=(0, LBP_POINTS+2))

    # GLCM
    glcm_features = []
    for d in GLCM_DISTANCES:
        glcm = graycomatrix(gray, distances=[d], angles=GLCM_ANGLES, symmetric=True, normed=True)
        for prop in ['contrast', 'homogeneity', 'energy', 'correlation']:
            glcm_features.extend(graycoprops(glcm, prop).flatten())
    # contrast = graycoprops(glcm, 'contrast').flatten()
    # # dissimilarity = graycoprops(glcm, 'dissimilarity').flatten()
    # energy = graycoprops(glcm, 'energy').flatten()
    # homogeneity = graycoprops(glcm, 'homogeneity').flatten()
    # correlation = graycoprops(glcm, 'correlation').flatten()
    # glcm_features  = np.hstack([contrast, energy, homogeneity, correlation])

    hybrid_features = np.concatenate([hog_features, hist_lbp, glcm_features])

    return hybrid_features
```

```python
# load dataset
def load_image(dataset_path):
    file_path = []
    labels = []

    path_imag = os.listdir(dataset_path)
    for index, categ in enumerate(path_imag):
        print(f"Loading Category {categ}")
        data = os.path.join(dataset_path, categ)
        for file in os.listdir(data):
            original_path = os.path.join(data, file)
            img = imread(original_path)

            # original image
            features_ori = proses_image(img)
            file_path.append(features_ori)
            labels.append(index)

            # terapkan augmentasi dan ekstraksi
            for _ in range(AUGMENT_FACTOR):
                img_aug = augment_image(img)
                features_aug = proses_image(img_aug)
                file_path.append(features_aug)
                labels.append(index)

        print(f"Loaded category {categ} is successfully")

    return np.array(file_path), np.array(labels)
X, y = load_image(DATASET_CITRA)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
```

```python
model = svm.SVC(class_weight='balanced')
model.fit(X_train, y_train)
```

```python
# best_svm = best_svm.best_estimator_
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy * 100:.2f}%')
```
