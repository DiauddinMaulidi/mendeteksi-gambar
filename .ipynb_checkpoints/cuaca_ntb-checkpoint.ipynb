```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../data/cuaca_ntb.csv')

data
```

```python
colomn_name = np.array(['id_lokasi', 'waktu', 'suhu_udara_min', 'suhu_udara_max', 'kelembapan_min', 'kelembapan_max', 'kelembapan_udara', 'suhu_udara', 'cuaca', 'arah_angin', 'kecepatan_angin'])
colomn_name
```

```python
row = np.array(data, dtype=np.str_)
row = np.char.replace(row, ";", ",")
row = row.flatten()
```

```python
new_data = [datas.split(',') for datas in row]
new_data
```

```python
df = pd.DataFrame(new_data, columns=colomn_name)
```

```python
df = df.drop(columns=['suhu_udara_min', 'suhu_udara_max', 'kelembapan_min', 'kelembapan_max'])
```

```python
df[['date', 'jam']] = [b.split(' ') for b in df['waktu']]
```

```python
df = df.drop(columns='waktu')
```

```python
df[['tahun', 'bulan', 'tanggal']] = [date.split('-') for date in df['date']]
```

```python
df = df.drop(columns='date')
```
